```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data_set=pd.read_csv("Downloads/company_sales_data.csv")
print(data_set)
```

```python
print(data_set.shape)
```

```python
print(data_set.info())
```

```python
total_profit=data_set["total_profit"].sum()
print("Total_profit from month 1 to 12:",total_profit)
```

```python
product_sales=data_set[["facecream", "facewash", "toothpaste", "bathingsoap", "shampoo" ,"moisturizer"]].sum()
print("product_wise_total:\n",product_sales)
```

```python
plt.figure(figsize=(6,4))
plt.plot(data_set["month_number"],data_set['facecream'],label="facecream",linestyle="--",marker='o')
plt.plot(data_set["month_number"],data_set['facewash'],label="facewash",linestyle="--",marker='o')
plt.plot(data_set["month_number"],data_set['toothpaste'],label="toothpaste",linestyle="--",marker='o')
plt.plot(data_set["month_number"],data_set['bathingsoap'],label="bathingsoap",linestyle="--",marker='o')
plt.plot(data_set["month_number"],data_set['shampoo'],label="shampoo",linestyle="--",marker='o')
plt.plot(data_set["month_number"],data_set['moisturizer'],label="moisturizer",linestyle="--",marker='o')
plt.xlabel("product")
plt.ylabel("Year sales")
plt.title("Sales")
plt.legend()
plt.grid()
plt.show()

```

```python
product_sales.plot(kind="bar",figsize=(6,4),color="red")
plt.title("Sales Graph")
plt.xlabel("product")
plt.ylabel("sales")
plt.xticks(rotation=45)
plt.grid(axis="y")
plt.show()
```

```python
plt.figure(figsize=(5,4))
plt.pie(product_sales,labels=product_sales.index,autopct="%1.1f%%",startangle=90)
plt.tight_layout()
plt.show()
```
