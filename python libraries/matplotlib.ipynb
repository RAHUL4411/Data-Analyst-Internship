```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
y=[2,4,6,8]
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel("Time(days)")
plt.ylabel("Distance(km)")
plt.title("Distance Over time")
plt.show()
```

```python
plt.plot(x,y,linestyle="-",color="lightblue",marker="*")
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.xlabel("x")
plt.ylabel("y")
plt.title("graph")
plt.plot(x,y,color="green",linestyle="--",marker="o")
plt.show()
```

```python
#bar chart
fruits=["Apple","Orange","Grapes","Banana"]
sales=[20,30,25,40]
plt.figure(figsize=(4,5))
plt.xlabel("fruits")
plt.ylabel("Sales")
plt.title("Sales of fruits")
plt.bar(fruits,sales,color="blue")
plt.show()
```

```python
#pie chart
size=[10,20,30,40]
names=["A","B","C","D"]
plt.pie(size,labels=names,autopct='%1.1f%%')
plt.show()
```

```python
#changing color and angle
size=[10,20,30,40]
names=["A","B","C","D"]
col=('#FF0000','#00FF00','#0000FF','#FFFF00') 

plt.pie(size,labels=names,colors=col,autopct='%1.1f%%',startangle=60)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sizes=[15,30,45,10]
labels=['category A','category B','category C','category D']
colors=sns.color_palette('rocket',len(sizes))
plt.pie(sizes,labels=labels, colors=colors,autopct='%1.1f%%',startangle=90)
plt.show()
```

```python
#scatter plot
x=[5,7,8,7,6,9,5,7]
y=[90,88,76,79,89,100,77,85]
plt.scatter(x,y,color='blue')
plt.show()
```

```python
fruits=['banana','mango','cherry']
sales=[10,15,11]
plt.barh(fruits,sales,color="purple")
plt.show()
```

```python
data=[20,11,30,33,6,28,55,46,70,66]
plt.hist(data,bins=5,color="skyblue",edgecolor="red")
plt.show()
```

```python
data=[20,11,30,33,6,28,55,46,70,66]
plt.boxplot(data)
plt.show()
```
