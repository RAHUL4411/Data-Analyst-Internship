```python

from collections import Counter
import ast
# Function to extract genre names from the JSON-like strings, handling potential errors
def extract_genre_names(genres_str):
    try:
        genres = ast.literal_eval(genres_str)
        return [genre['name'] for genre in genres]
    except (SyntaxError, ValueError):
        return ['Unknown']  # Replace with a default value or handle error as needed

# Apply the function to the 'genres' column
df['genres_list'] =df['genres'].apply(extract_genre_names)

# Flatten the list of genres and count occurrences
all_genres = [genre for sublist in df['genres_list'] for genre in sublist]
genre_counts = Counter(all_genres)

# Convert the counter to a DataFrame for easy plotting
genre_counts_df = pd.DataFrame(genre_counts.items(), columns=['Genre', 'Count'])
```

## Observation: Genre Distribution

- Extracted genre names from the `genres` column using `ast.literal_eval()` and stored them in a new column `genres_list`.
- Flattened the nested genre lists and counted occurrences using `collections.Counter`.
- Converted the genre frequency data into a DataFrame `genre_counts_df` for easy plotting.
- This process reveals the most common genres in the dataset, enabling genre-based analysis and visualization.
- Helps identify dominant themes (e.g., Action, Drama, Comedy) and supports filtering or trend exploration.

```python
print("Genre counts:")
for genre,count in genre_counts.items():
    print(f"{genre}:{count}")
```

## Observation: Genre Frequency Summary

- Printed the frequency of each genre using a loop over `genre_counts`.
- This gives a clear view of how many movies fall under each genre.
- Common genres like **Drama**, **Comedy**, and **Action** typically appear at the top.
- Less frequent genres (e.g., **Western**, **War**, **Music**) help identify niche categories in the dataset.
- This breakdown supports genre-based filtering, trend analysis, and targeted visualizations.

```python
a=genre_counts_df.plot(kind="bar",x="Genre",y="Count",color="pink")
plt.xlabel("genre")
plt.ylabel("counts")
plt.title("Genre count")
for index,value in enumerate(genre_counts_df['Count'] ):
    plt.text(index,value,str(value),ha="center",va="bottom",fontsize=10,fontweight="bold")
plt.show()
```

###  Observation: Genre Count Distribution

- The bar chart displays the frequency of each genre in the dataset.
- **Drama**, **Comedy**, and **Action** are among the most represented genres, indicating their dominance in movie production.
- Less frequent genres like **War**, **Western**, and **Music** appear with smaller bars, highlighting niche categories.
- Annotated counts on each bar improve readability and help quickly compare genre popularity.
- This visualization supports genre-based filtering, trend analysis, and helps prioritize genres for deeper exploration.
